```python
import pandas as pd
import time
from sodapy import Socrata

# Unauthenticated client only works with public data sets. Note 'None'
client = Socrata("data.cityofnewyork.us", app_token = 'DZ7Ww3ROSOeUk8rOmGlJRr99A')
```

```python
def fetch_paginated_data(client, dataset_id, where_clause, limit=100000):
    offset = 0
    all_data = []

    while True:
        results = client.get(
            dataset_id,
            where=where_clause,
            limit=limit,
            offset=offset
        )
        if not results:
            break
        all_data.extend(results)
        offset += limit
        time.sleep(0.1)

    return pd.DataFrame.from_records(all_data)

# Usage
final_df = fetch_paginated_data(client, "rbx6-tga4", "issued_date >= '2025-02-01T00:00:00.000'")
```
